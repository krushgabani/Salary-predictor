```python
import pandas as pd
import joblib

model = joblib.load('Salary_prediction.pkl')
columns = joblib.load('columns.pkl')

# Example user input
user_input = {
    'experience_years': 17,
    'skills_count': 7,
    'certifications': 1
}

# One-hot encode manually — same way get_dummies(drop_first=True) works
def ohe(input_dict, prefix, all_vals, chosen):
    cats = sorted(all_vals)[1:]  # drop_first removes first alphabetically
    for c in cats:
        input_dict[f"{prefix}_{c}"] = 1 if chosen == c else 0

ohe(user_input, 'job_title',       ['AI Engineer','Business Analyst','Cloud Architect','Cybersecurity Analyst','Data Analyst','Data Engineer','Data Scientist','DevOps Engineer','Frontend Developer','Full Stack Developer','Machine Learning Engineer','Mobile Developer','Network Engineer','Product Manager','QA Engineer','Software Engineer','UX Designer'], 'Cybersecurity Analyst')
ohe(user_input, 'education_level', ['Bachelor','High School','Master','PhD'], 'PhD')
ohe(user_input, 'industry',        ['Aerospace','Agriculture','Automotive','Banking','Consulting','Defense','Education','Energy','Entertainment','Finance','Government','Healthcare','IT Services','Insurance','Legal','Logistics','Manufacturing','Marketing','Media','Pharmaceuticals','Real Estate','Retail','Telecom','Transportation'], 'Finance')
ohe(user_input, 'company_size',    ['Large','Medium','Small','Startup'], 'Medium')
ohe(user_input, 'location',        ['Australia','Brazil','Canada','China','France','Germany','India','Japan','Mexico','Netherlands','Singapore','South Korea','Sweden','UAE','UK','USA'], 'India')
ohe(user_input, 'remote_work',     ['Hybrid','No','Yes'], 'No')

# Align with training columns
df_input = pd.DataFrame([user_input])
df_input = df_input.reindex(columns=columns, fill_value=0)

salary = model.predict(df_input)
print("Predicted Salary:", salary[0])
```
